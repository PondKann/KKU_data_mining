# Box plot

```
from matplotlib import pyplot as plt
```

นำแกน x มาสร้าง boxplot

```
plt.boxplot(data['x'])
```

ใส่ vert เพื่อในกล่องอยู่ในแนวนอน

```
plt.boxplot(data['x'],vert=False)
```

เปลี่ยน whisker เป็น 1.2 ให้มี ค่า out lier ให้มีมากกว่าเดิม 

```
plt.boxplot(data['x'],vert=False,whis=1.2)
```

```
oput = plt.boxplot(data['x'],vert=False,whis=1.2)
```

ดูค่าที่มัน มากกว่า 1.2IQR (ค่า outlier)

```
oput['fliers'][0].get_xdata()
```

เส้น whisker วาดจากค่า Q1(=44.1026) ถึง 1.2IQR(=22.3077) (1.2 เรากำหนดใหม่ ปกติเป็น 1.5)

```
oput['whiskers'][0].get_xdata()
```

```
oput['whiskers'][1].get_xdata()
```

# Scatter plot

```
plt.scatter(data['x'],data['y'])
```

```
plt.scatter(data['x'],data['y'],s=25)
```

ขนาดของจุดเป็นไปตามค่า y ยิ่งค่า y มาก ขนาดจะยิ่งใหญ่

```
plt.scatter(data['x'],data['y'],s=data['y'])
```

เพิ่ม c = จุดเป็นสีต่าง ๆ 

```
plt.scatter(data['x'],data['y'],s=data['y'],c='m')
```

```
plt.scatter(data['x'],data['y'],s=data['y'],c='m',marker='*')
```

```
plt.scatter(data['x'],data['y'],s=data['y'],c='m',marker='*',alpha=0.4)
plt.scatter([44,42,50],[15,14,80],s=50,alpha=0.6)
```
